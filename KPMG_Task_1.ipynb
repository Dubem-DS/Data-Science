# Customer Demographic

```python
raw_data_CD = pd.read_excel(r'C:\Users\user\Downloads\.ipynb_checkpoints\KPMG_VI_New_raw_data_update_final.xlsx', sheet_name= 'CustomerDemographic')
```

```python
raw_data_CD.isnull().sum()
```

```python
raw_data_CD_no_mv = raw_data_CD.dropna(axis=0)
```

```python
raw_data_CD_no_mv.isnull().sum()
```

```python
raw_data_CD_no_mv['deceased_indicator'].value_counts()
```

```python
raw_data_CD_no_mv
```

```python
from datetime import date
now = pd.Timestamp('now')
raw_data_CD_no_mv['DOB'] = pd.to_datetime(raw_data_CD_no_mv['DOB'], format= '%y%m%d')
raw_data_CD_no_mv['DOB'] = raw_data_CD_no_mv['DOB'].where(raw_data_CD_no_mv['DOB'] < now, raw_data_CD_no_mv['DOB'] - np.timedelta64(100, 'Y'))
raw_data_CD_no_mv['age'] = (now - raw_data_CD_no_mv['DOB']).astype('<m8[Y]')
raw_data_CD_no_mv.head()
```
